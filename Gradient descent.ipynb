### 選用的函式為 x**4 - 0.2x**3 + 3x**2 - 4x - 5

```python
def f(x):
    return x**4 - 0.2*x**3 + 3*x**2 - 4*x - 5
def df(x):#define the function being devirated
    return 4*x**3 - 6*x**2 + 6*x - 4 
```

### Gradient descent的函式形式

```python
def gradient_descent(x):
    x = x - learning_rate * df(x)
    #print(x)
    return x
```

### 可以選定"start point" 與 "learning rate"

```python
learning_rate = 0.001
start_pt = 5
#初始化 count
count = 0
#宣告 x space
x_space = np.linspace(-5,5,1000)
```

```python
x = start_pt
count = 0
while(learning_rate * df(x) > 0.001):
    x = gradient_descent(x)
    plt.plot(x_space,f(x_space),c = 'blue')
    plt.scatter(x,f(x),s=20,c = 'red')
    count += 1
    print('iterate time : %s'%count   ,'temporary x_value : %s'%x)
print('In summary:')
print('-------------------------------------')
print("Total count = ",count)
print("Final x = %s"%x)
```
